# 2.0. Edit 'Sunshine hours (City)' column data

```python
data
```

```python
# If we analyze the data in the column 'Sunshine hours (City)' we can notice that we have data '-' at index 19.
# Also in this column the data type is object (string)
# The task is to modify the data '-', and then to convert the column data type from object to int
data.dtypes
```

```python
# The Sunshine hours (City) column is an object data type, we want to convert it to int 64
# The problem with the conversion is the '-' sign on the 19 index
data['Sunshine hours(City)']
```

```python
# This is the Pandas Series
type(data['Sunshine hours(City)'])
```

```python
# We change the character '-' with the number 0
data['Sunshine hours(City)'].replace(to_replace = '-', value = 0, inplace = True)
```

```python
# Let's check the results
data['Sunshine hours(City)']
```

```python
data.dtypes
```

```python
# Changing the data type of the column 'Sunshine hours (City)'
data['Sunshine hours(City)'] = data['Sunshine hours(City)'].astype('int64')
```

```python
# We verified that the data type has changed
data.dtypes
```

```python
# Data on index 19, we want to replace with the average number of all other rows
round(data.loc[(data['Sunshine hours(City)'] != 0) == True, 'Sunshine hours(City)'].mean())
```

```python
# We change the data on index 19, with an average
data.loc[(data['Sunshine hours(City)'] == 0) == True, 'Sunshine hours(City)'] = round(data.loc[(data['Sunshine hours(City)'] != 0) == True, 'Sunshine hours(City)'].mean())
```

```python
# Let's check the results
data
```

# 6. Create a new DataFrame column 'Life Expectancy Range'

```python
data.sort_values(by='index', inplace=True)
data
```

```python
# For the purpose of creating a new column 'Life Expectancy Range' we create the function LifeExpectancyRange
def LifeExpectancyRange(year):
    if year>=80:
        return 'Above 80'
    elif year< 80 and year>=70:
        return 'Between 70 and 79'
    elif year<70 and year>=60:
        return 'Between 60 and 69'
    else:
        return 'Under 60'
```

```python
# In this step, we created a new column based on the LifeExpectancyRange function
data['Life Expectancy Range'] = data['Life expectancy(years) (Country)'].apply(lambda x: LifeExpectancyRange(x))
```

```python
data
```

```python
# Now we will do Pivot tables according to the index 'Life Expectancy Range', 'City', where we will do by groups
# show which cities belong to which categories 'Life Expectancy Range'
pd.pivot_table(data, 
               index=['Life Expectancy Range', 'City'], 
               values=['Life expectancy(years) (Country)'])
```
